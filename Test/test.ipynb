```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

backend = AerSimulator()
# backend = AerSimulator(device='GPU')
```

```python
job = backend.run(qc, shots=1000)
result = job.result()
```

```python
print(result.results[0].data)
```

```python
counts = result.get_counts()
print(counts)
```
